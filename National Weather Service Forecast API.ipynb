```python
def gets_temperature_forecast():
    import requests
    import json
    from geopy.geocoders import Nominatim
    from datetime import datetime
    
    #Ask user to enter location
    
    user_address = input('Where do you live?')
    
    geolocator = Nominatim(user_agent='my_address')
    location = geolocator.geocode(user_address)
    lat = location.latitude
    lon = location.longitude
   
    #Connection to weather api
    pointsurl = 'https://api.weather.gov/points/%(1)s,%(2)s'%{'1': lat, '2': lon}
    pointsresponse = requests.get(pointsurl)
    if pointsresponse.status_code == 200:
        pointdata = pointsresponse.json()
        
        mainprops = pointdata['properties']
        CWA = mainprops['cwa']
        GridX = mainprops['gridX']
        GridY = mainprops['gridY']
        
        gridpointsurl = 'https://api.weather.gov/gridpoints/%(3)s/%(4)s,%(5)s' %{'3': CWA, '4': GridX, '5': GridY}
        gridresponse = requests.get(gridpointsurl)
        
        if gridresponse.status_code == 200:
            griddata = gridresponse.json()
            
        maingridprops = griddata ['properties']
        temperatures = maingridprops ['temperature']
        tempvalues = temperatures ['values']
        
        for value in tempvalues:
            time = value ['validTime'].rsplit('+')[0]
            
            datetimeobj = datetime.strptime(time, '%Y-%m-%dT%H:%M:%S')
            
            tempC = value['value']
            tempF = tempC * 9 / 5 + 32
            
            degreeSign = u'\N{DEGREE SIGN}'
            print('Date: {0} Time: {1} Temperature: {2: .2f}{3}C ({4: .1f}{3}F)'.format(datetimeobj.date(), datetimeobj.time(), tempC, degreeSign, tempF))
            
            
```

```python
gets_temperature_forecast()
```
